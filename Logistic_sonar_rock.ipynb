# Importing Libraries

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score,classification_report,confusion_matrix
```

# Features and target selection and splitting data

```python
# Features or independent variables
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model Training

```python
logist_reg=LogisticRegression()
```

```python
logist_reg.fit(X_train,y_train)
```

# Model evaluation

```python
y_pred=logist_reg.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## Comparing Model performance with Training data

```python
print(classification_report(y_train,logist_reg.predict(X_train)))
```

#### So, there is a more than 10% difference between accuracies of training and testing data
#### Clearly our model is overfitting

# GridSearchCV

We'll be using Grid Search for finding the best parameters for Logistic Regression

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier=LogisticRegression(max_iter=200)
```

```python
param_grid={'penalty':['l1','l2','none'],
           'C':[0.1,1,10],
           'solver':['newton-cg','lbfgs','liblinear','sag']}
```

```python
grid=GridSearchCV(classifier,param_grid=param_grid,scoring='accuracy',cv=10,n_jobs=-1)
```

```python
grid.fit(X,y)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
grid.cv_results_
```

We put the results of Gridsearchcv in a dataframe for readability 

```python
grid_results=pd.DataFrame(grid.cv_results_)
```

```python
grid_results.head()
```

```python
grid_results[['params','mean_test_score','std_test_score']].sort_values(by='mean_test_score',ascending=False)
```
